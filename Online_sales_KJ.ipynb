### Business Information

**About Dataset**

This dataset provides a comprehensive overview of online sales transactions across different product categories. Each row represents a single transaction with detailed information such as the order ID, date, category, product name, quantity sold, unit price, total price, region, and payment method.

**Columns:**

* **Order ID:** Unique identifier for each sales order.
* **Date:** Date of the sales transaction.
* **Category:** Broad category of the product sold (e.g., Electronics, Home Appliances, Clothing, Books, Beauty Products, Sports).
* **Product Name:** Specific name or model of the product sold.
* **Quantity:** Number of units of the product sold in the transaction.
* **Unit Price:** Price of one unit of the product.
* **Total Price:** Total revenue generated from the sales transaction (Quantity * Unit Price).
* **Region:** Geographic region where the transaction occurred (e.g., North America, Europe, Asia).
* **Payment Method:** Method used for payment (e.g., Credit Card, PayPal, Debit Card).

URL: https://www.kaggle.com/datasets/shreyanshverma27/online-sales-dataset-popular-marketplace-data/data

1. Identify top-performing regions and product categories to inform regional sales strategies and inventory management.

```python
# Group by Region and Product Category, then sum the Total Revenue
df_grouped_rp = df.groupby(['Region', 'Product Category'])['Total Revenue'].sum().reset_index()

# Calculate total overall revenue
total_revenue = df['Total Revenue'].sum()

# Calculate proportion from total
df_grouped_rp['Percentage'] = (df_grouped_rp['Total Revenue'] / total_revenue * 100).round(1)

# Sort by Region first and then by original Total Revenue in descending order
df_sorted_rp = df_grouped_rp.sort_values(['Region', 'Total Revenue'], ascending= [True, False])

print("\nRevenue Analysis by Region and Product Category:")
print(df_sorted_rp[['Region', 'Product Category', 'Total Revenue', 'Percentage']])
```

```python
# Pivot the data to get each Product Category as a separate column
df_pivot = df_sorted_rp.pivot(index='Region', columns='Product Category', values='Total Revenue').fillna(0)

# Sort columns within each row (region) by total revenue
df_pivot = df_pivot[df_pivot.sum().sort_values(ascending=False).index]

df_pivot
```

```python
# Plotting a stacked bar chart with custom colors
ax = df_pivot.plot(kind='bar', stacked=True, figsize=(12, 8), colormap="cividis")

# Annotate each bar with both value and percentage
for container in ax.containers:
    # Calculate total revenue per region for percentage calculation
    totals = df_pivot.sum(axis=1)
    for i, bar in enumerate(container):
        # Get the height of each bar (the sales value)
        height = bar.get_height()
        # Calculate the percentage
        percentage = height / totals[i] * 100
        # Only add annotation if value is greater than 0
        if height > 0:
            ax.annotate(f'{height:,.0f}\n({percentage:.1f}%)',  # Show value and percentage on separate lines
                        xy=(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2),
                        ha='center', va='center', color='black', fontsize=9)  # Reduced font size slightly

# Labels and title
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Region and Product Category (Sorted)')
plt.legend(title='Product Category')
plt.show()
```

**Insight:**

Regional Revenue: The North America region generates the most revenue, accounting for almost 44% of the total.
Product Category Preference: However, purchasing behavior varies significantly across regions. While North America primarily favors Electronic products, Europe leans towards Home Appliances, and Asia shows a preference for Sports products.

**Recommendation:**

To optimize inventory and marketing strategies, the company should:

* Stock up on Electronic products in the North America region.
* Increase the stock of Home Appliances in Europe.
* Prioritize Sports products in Asia.

By tailoring inventory and marketing efforts to regional preferences, the company can enhance customer satisfaction and drive sales.

6. Understand the revenue share from each product category to support category-level budgeting, pricing, and promotional decisions.

```python
# Group by 'Product Category' and calculate the total revenue and units sold per category

sale_cate = df.groupby('Product Category')[['Total Revenue', 'Units Sold']].sum().reset_index()

# Sort by 'Total Revenue' in descending order
sale_cate = sale_cate.sort_values('Total Revenue', ascending=False)

# Calculate the percentage of total revenue for each category
total_revenue = df['Total Revenue'].sum()  # Ensure total_revenue is defined
sale_cate['Percentage'] = (sale_cate['Total Revenue'] / total_revenue * 100).round(2)

# Calculate the Average order Unit Sold for each category
sale_cate['No. of Orders'] = df.groupby('Product Category')['Transaction ID'].count().reset_index(drop=True)
sale_cate['Avg unit order per category'] = (sale_cate['Units Sold'] / sale_cate['No. of Orders']).round(2)

# Display the final DataFrame
sale_cate
```

**Insight:**

**Revenue Contribution by Product Category**

* Electronics and Home Appliances are the primary revenue drivers, contributing 43.42% and 23.14%, respectively, to the total revenue. Together, these two categories account for over 66% of the total revenue.

* Beauty Products and Books have the lowest revenue shares at 3.25% and 2.31%, respectively. These categories may require targeted promotions to boost visibility and sales. Alternatively, if profit margins are low, reducing inventory levels for these categories could be considered.

**Sales Volume (Units Sold) and Average Units per Order**

* Clothing has a moderate share but a large number of Units Sold (145), meaning it likely sells in higher quantities at a lower price point.

* Sports and Books categories also show relatively high average units per order, at 2.20 and 2.85, respectively, indicating that customers often purchase multiple items from these categories.

**Order Behavior and Customer Engagement**

* Electronics has high revenue but a relatively low average units per order (1.65), suggesting a focus on higher-value. 

* In contrast, Clothing and Books are more volume-driven, with higher average unit counts per order, indicating a focus on lower-quantity items.

**Recommendations**

* Promotional Spend: Allocate more promotional budget to Electronics and Home Appliances, as they yield the highest revenue impact.
* Bundling and Upselling: Consider bundling strategies for Clothing and Books, where there’s potential to leverage higher purchase quantities.
* Customer Loyalty Programs: For categories with frequent purchases, such as Sports and Clothing, a rewards program could encourage repeat purchases and build loyalty.

8. Evaluate the impact of unit pricing on profitability across regions, helping adjust pricing strategies where necessary.

```python
# Grouping data by Region and Product Category to calculate Total Revenue
reg_price = df.groupby(['Region', 'Product Category'])['Total Revenue'].sum().reset_index()

# Calculating the average unit price for each group
reg_price['Avg_unit_Price'] = (
    df.groupby(['Region', 'Product Category'])['Unit Price'].sum() /
    df.groupby(['Region', 'Product Category'])['Units Sold'].sum()
).reset_index(drop=True)

reg_price
```

**Insight:**

* North America leads in revenue from Electronics, generating $34,982 with a high average unit price of $419. This suggests that customers in this region prioritize high-quality, premium products and are less concerned with the higher price if the product meets their quality expectations.
* Europe shows a strong focus on premium categories, particularly Home Appliances, with revenue of $18,646 and the second-highest average unit price of $217. This indicates a preference for high-value products in this region.
* Asia generates substantial revenue in Sports at $14,326.52, with an average unit price of $118.77, indicating a market for mid-priced sports equipment or apparel.
* Clothing in Asia also performs reasonably well with $8,128.93 in revenue, accompanied by an affordable average unit price of $18.63, suggesting a focus on budget-conscious consumers.

**Conclusion:**

* The data indicates that North America and Europe are more focused on high-end, premium products, particularly in categories like Electronics and Home Appliances.
* Asia, on the other hand, presents a more balanced market, with both affordable clothing and mid-priced sports products, likely appealing to a broader, price-sensitive consumer base.
